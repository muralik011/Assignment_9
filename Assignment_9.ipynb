<b>1. To what does a relative path refer?

Relative path refers to the address referred from the current working directory. For example if C:\dir is current working directory the relative path of C:\dir\dir1\dir2 will be dir1\dir2

<b>2. What does an absolute path start with your operating system?

Drive/partition label like C:, D:, etc

<b>3. What do the functions os.getcwd() and os.chdir() do?

os.getcwd() returns the absolute path of current working directory. os.chdir() changes the current working directory to a specified path

<b>4. What are the . and .. folders?

. refers to current working directory, while .. refers to parent directory of current working directory

dir name: eggs<br>dir path: C:\bacon\eggs\ <br>base name: spam.txt

The existing content of the file is truncated and repaced with newly written content
